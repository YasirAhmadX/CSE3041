# Pandas Visualizations  

Data Visualization is a technique used for the graphical representation of data. (Eg.: scatterplot, line, bar, piechart, histrogram).  

### Libraries 
* <u>**pandas**</u> : tabular data manipulation
* <u>**matplotlib**</u> : simple graphs like lines,scatterplots,etc.
* **seaborn** : more complex plot, like correlation plots.
* **ggplot2** : Grammar of Graphics. Can be used along with R

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('plotdata.csv')
df
```

## Scatter plot


```python
df.plot(x='Unemployment_rate',y='Index_price', kind='scatter')
plt.show()
```

## Line chart

```python
df.plot(x='Year',y='Unemployment_rate', kind='line')
plt.title("Line chart")
plt.show()
```

## Bar chart

```python
df.plot(x='Year',y='Index_price', kind='bar',grid = True)
plt.show()
```

## Pie chart

```python
df.plot(x = 'Unemployment_rate',y='Year', kind='pie')
plt.show()
```

## Histrogram

```python
df.plot(x='Index_price',y='Unemployment_rate',kind ='hist')
```
