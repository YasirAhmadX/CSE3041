## Task


```python
df = pd.DataFrame({'name':['alpha','beta','charlie','delta','epsilon','foxbat','gamma','henry','iota','jack'],
                  'DoB':['8/10/2003','4/12/2003','18/1/2003','1/1/2003','7/1/2003','26/8/2003','8/10/2003',
                        '28/2/2003','5/11/2002','9/4/2003']})
df
```

```python
df['DoB'] = pd.to_datetime(df['DoB'])
df
```

```python
df['year'] = df['DoB'].dt.year
df['month'] = df['DoB'].dt.month
df['day'] = df['DoB'].dt.day
df
```

```python
today = pd.to_datetime('today')
df['age'] = today.year - df['DoB'].dt.year
df
```

### Set date as index

```python
df=df.set_index(['DoB'])
df
```

```python
#students born in 2003
df.loc['2003']
```

```python
df.loc['2003','age'].sum()
```
