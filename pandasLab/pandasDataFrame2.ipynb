# <u> Sorting</u>

Sort a data fram according to any column using  `df.sort_values()`

```python
df.sort_values(by='record_high',ascending=False)
```

* **<u> Sorting row wise</u>**  
    *`df.sort_index(axis=0)` will sort according to y axis  
    *`df.sort_index(axis=1)` will sort according to x axis(column name)

```python
df.sort_index(axis=0)
```

```python
df.sort_index(axis=1)
```
